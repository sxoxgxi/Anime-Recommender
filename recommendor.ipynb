```python
selected_anime_df['tags'] = selected_anime_df['Synopsis']+selected_anime_df['Themes']+selected_anime_df['Genres']
```

```python
new_df = selected_anime_df[['tags', 'Title', 'ID']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = [ps.stem(i) for i in text.split()]
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# consumes memory 
similarity = cosine_similarity(vectors)
```

```python
def recommend(anime):
    anime_index = new_df[new_df['Title']== anime].index[0]
    distances = similarity[anime_index]
    anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in anime_list:
        print(new_df.iloc[i[0]].Title)
```

```python
recommend("Death Note")
```
